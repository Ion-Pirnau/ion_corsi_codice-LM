```python
variabili_numeriche = list(df.columns[1:-1])
print(variabili_numeriche)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

#df_n = pd.to_numeric(df[variabili_numeriche])
lista_val = df[variabili_numeriche].values.tolist()
lista_pulita = [[val.strip().replace('"', '') for val in sublist] for sublist in lista_val]
lista_pulita

df_n = pd.DataFrame(lista_pulita)
#df_n

scaled_data = scaler.fit_transform(df_n)
scaled_data
```

```python
import matplotlib.pyplot as plt

pca = PCA().fit(scaled_data)

plt.plot(range(1, pca.n_components_ + 1), pca.explained_variance_ratio_, marker='o')
plt.xlabel('Componenti della PCA')
plt.ylabel('Varianza speigata')
plt.title('Risultati della PCA')
plt.show()
```

```python
pca = PCA(n_components=2).fit(scaled_data)
pca_data = pca.transform(scaled_data)
```

```python
print(pca.explained_variance_ratio_)
```

```python
unique_labels = np.unique(df[' "Competition"'])
lista_unique_labels = [val.strip().replace('"', '') for val in unique_labels]

lista = df[' "Competition"'].tolist()

lista_competition = [val.strip().replace('"', '') for val in lista]
print(lista_competition)
print(unique_labels)
print((df[' "Competition"'] == ' "Decastar"').values)
```

```python
colors = plt.cm.tab20(np.linspace(0, 1, len(lista_unique_labels)))
label_to_color = dict(zip(lista_unique_labels, colors))
print(label_to_color)

fig, ax = plt.subplots()
mask = (df[' "Competition"'] == ' "Decastar"').values
ax.scatter(pca_data[mask, 0], pca_data[mask, 1], color=label_to_color['Decastar'], label='Decastar')
    
mask = (df[' "Competition"'] == ' "OlympicG"').values
ax.scatter(pca_data[mask, 0], pca_data[mask, 1], color=label_to_color['OlympicG'], label='OlympicG')


ax.set_xlabel('Componente 1')
ax.set_ylabel('Componente 2')

ax.legend()

plt.show()
```

```python
print(abs(pca.components_))
```

```python
ax = sns.heatmap(pca.components_,
                 cmap='YlGnBu',
                 yticklabels=[ "PCA"+str(x) for x in range(1, pca.n_components_+1)],
                 xticklabels=list(variabili_numeriche),
                 cbar_kws={"orientation": "vertical"})

ax.set_aspect("equal")
```

```python
loadings = pca.components_
num_pc = pca.n_features_in_
pc_list = ["PC"+str(i) for i in list(range(1, num_pc+1))]
loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
loadings_df

loadings_df['variable'] = variabili_numeriche
loadings_df = loadings_df.set_index('variable')
loadings_df

ax = sns.heatmap(loadings_df, annot=True, cmap='Spectral')
plt.show()
```

```python
pcs = pca.components_
fig = plt.figure(figsize=(6,5))
ax = fig.add_subplot(1, 1, 1)
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])

for i, (x,y) in enumerate(zip(pcs[0, :], pcs[1, :])):
    ax.plot([0, x], [0, y], color='g')
    ax.text(x, y, df.columns[i], fontsize='15')
```

```python
var_num_iris = list(df_iris.columns[0:-1])
print(var_num_iris)
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_iris[var_num_iris])
scaled_data
```

```python
pca = PCA().fit(scaled_data)

# Crea un grafico della varianza spiegata per ogni componente
plt.plot(range(1, pca.n_components_ + 1), pca.explained_variance_ratio_, marker='o')
plt.xlabel('Componenti della PCA')
plt.ylabel('Varianza spiegata')
plt.title("Risultati della PCA")
plt.show()
```

```python
pca = PCA(n_components=2).fit(scaled_data)
pca_data = pca.transform(scaled_data)
print(pca.explained_variance_ratio_)

eigenvalues = pca.explained_variance_
n_samples = scaled_data.shape[0]
cov_matrix = np.dot(scaled_data.T, scaled_data) / n_samples
for eigenvalue, eigenvector in zip(eigenvalues, pca.components_):
    print(np.dot(eigenvector.T, np.dot(cov_matrix, eigenvector)))
    print(eigenvalue)
```

```python
unique_labels = np.unique(df_iris["target"])
colors = plt.cm.tab20(np.linspace(0, 1, len(unique_labels)))
label_to_color = dict(zip(unique_labels, colors))

# genera il grafico
fig, ax = plt.subplots()
for label in unique_labels:
    mask = (df_iris["target"] == label).values
    ax.scatter(pca_data[mask, 0], pca_data[mask, 1], color=label_to_color[label], label=label)

ax.set_xlabel('Componente 1')
ax.set_ylabel('Componente 2')

ax.legend()

plt.show()
```
